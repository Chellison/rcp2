```python
out_bg = pd.read_csv(r'C:\Users\cle9a\Documents\rcp2_8_2_2022\Data\Model Outputs\CombinedModels-FileStructureColumns-pop0-2022-08-24.csv')
out_bg
```

```python
col_mode = ['state', 'NFIRS_start', 'NFIRS_end', 'geography', 'state_id', 'county_id']
col_sum = [ 'total_fires_reported', 'total_fires_reported_sev','recorded_home_visits', 'tot_population','house_tot_occ_cnt']
col_div = [ 'years_fires_reported', 'max_fires_1k', 'median_fires_1k', 'fire_severity_risk_score', 'fire_risk_score', 'smoke_detector_risk_frac',
       'smoke_detector_risk_score','SVI_Theme_1','CombinedRiskScore', 'did_not_work_past_12_mo', 'house_pct_live_alone',
       'pct_alt_heat', 'house_yr_pct_before_1960', 'inc_pct_poverty','race_pct_black_or_amind','inc_pcincome', 'inc_hhmean', 
        'race_pct_white', 'race_pct_whitenh','race_pct_whiteh', 'race_pct_nwhitenh', 'race_pct_hisp', 'race_pct_black', 
        'race_pct_amind', 'race_pct_asian', 'race_pct_napi', 'race_pct_other', 'race_pct_two+', 'age_pct_under25', 
        'age_pct_over65','disability_pct', 'snap_past_12_mo','single_unit_home','multi_unit_home', 'mobile_home', 
        'boat_van_rv_home','spanish_speaking']
#other columns- GEOID, state_rank, an county_rank
out_bg.columns
```

```python
#get tract-level info
out_bg['tract'] = out_bg['GEOID'].str[:-1]
out_bg['tract']
```

```python
out_sum = out_bg.groupby('tract')[col_sum].sum()
print(out_sum.shape)
out_mode = out_bg.groupby('tract')[col_mode].agg(pd.Series.mode)
print(out_mode.shape)
out_bg_div = out_bg[col_div].mul(out_bg['tot_population'], axis = 0)
out_bg_div['tract'] = out_bg['tract']
out_div = out_bg_div.groupby('tract')[col_div].sum()
out_div = out_div.div(out_sum.tot_population, axis=0)
print(out_div.shape)
```

```python
out_tract = out_sum.merge(out_mode, how = 'left', right_index = True, left_index = True)
out_tract = out_tract.merge(out_div, how = 'left', right_index = True, left_index = True)
out_tract
```

```python
out_tract['state_rank'] = out_tract.groupby('state_id')['CombinedRiskScore'].rank(ascending=False)
out_tract['county_rank'] = out_tract.groupby('county_id')['CombinedRiskScore'].rank(ascending=False)
```

```python
out_name = f"CombinedModels-FileStructureColumns-tract-{date.today()}.csv"
out_tract.to_csv(utils.DATA['model-outputs'] / out_name)
```

```python
out_sum = out_bg.groupby('county')[col_sum].sum()
print(out_sum.shape)
out_mode = out_bg.groupby('county')[col_mode].agg(pd.Series.mode)
print(out_mode.shape)
out_bg_div = out_bg[col_div].mul(out_bg['tot_population'], axis = 0)
out_bg_div['county'] = out_bg['county']
out_div = out_bg_div.groupby('county')[col_div].sum()
out_div = out_div.div(out_sum.tot_population, axis=0)
print(out_div.shape)
```

```python
out_county = out_sum.merge(out_mode, how = 'left', right_index = True, left_index = True)
out_county = out_county.merge(out_div, how = 'left', right_index = True, left_index = True)
out_county
```
